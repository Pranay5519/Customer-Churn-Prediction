```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
pd.set_option('display.max_columns',None)
df = pd.read_excel(r'/content/CHURNDATA (1) (1).xlsx')
df.head()
```

## Missing Values

```python
#check missing values
df.isna().sum()
```

```python
# knn imputer
from sklearn.impute import SimpleImputer , KNNImputer
knn  = KNNImputer()
simple_imputer = SimpleImputer(strategy='most_frequent')
df['CUS_Month_Income'] = knn.fit_transform(df[['CUS_Month_Income']])
df['CUS_Gender'] = simple_imputer.fit_transform(df[['CUS_Gender']])
```

## Data Cleaning

```python
del df['CIF']
del df['CUS_DOB']
del df['CUS_Customer_Since']
#del df['CUS_Target']
```

```python
le_cols = ['CUS_Gender','Status'] # gender col - one hot
one_hot_cols  = ['CUS_Marital_Status']


#encoding
import numpy as np
def encoder(ordinal_cols =None , one_hot_cols = None,le_cols = None  , df=df):

    from sklearn.preprocessing import LabelEncoder , OneHotEncoder , OrdinalEncoder
    le = LabelEncoder()
    od = OrdinalEncoder()
    one_hot = OneHotEncoder(handle_unknown='ignore',sparse =False)
    if ordinal_cols:
        for x in ordinal_cols:
            df[x] = od.fit_transform(df[[x]])
        #print(od.categories_)
    if one_hot_cols:
        for x in one_hot_cols:
            data = one_hot.fit_transform(df[[x]])
            cols = np.array(one_hot.categories_)[0 ,:].tolist()


            enc_data = pd.DataFrame(data , columns = cols)
            del df[x]

            df = df.join(enc_data)
    if le_cols:
        for x in le_cols:
            df[x] = le.fit_transform(df[x])

    return df

df = encoder(one_hot_cols=one_hot_cols , le_cols=le_cols,df=df)
```

```python
#ordinal encoding
df['TAR_Desc'] = df['TAR_Desc'].map({'LOW':0 , 'MIDLE':1,'EXECUTIVE':3,'PLATINUM':2})
```

```python
import numpy as np

def detect_outliers_zscore(data, threshold=3):
    z_scores = (data - np.mean(data)) / np.std(data)
    outlier_indices = np.where(np.abs(z_scores) > threshold)[0]
    return outlier_indices


outlier_indices = detect_outliers_zscore(df['CUS_Month_Income'])
print(len(outlier_indices))
print(df.shape)
df = df.drop(outlier_indices)
print(df.shape)
```

# Automation

```python
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import MinMaxScaler
def train_model(model_name, preprocessing_types, algo_name ,X ,y , target_col_name , feature_selection = None ,random_state= None):
    feature_names = X.columns
    # Perform train-test split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,stratify = y_resampled , random_state = random_state ,)

    # Preprocessing

    ss = preprocessing_types
    X_train_transformed = ss.fit_transform(X_train)
    X_test_transformed = ss.fit_transform(X_test)

    if feature_selection:

        fs = feature_selection
        fs.fit(X_train_transformed , y_train)
        X_train_transformed = fs.transform(X_train_transformed)
        X_test_transformed = fs.transform( X_test_transformed)
        selected_feature_indices = fs.get_support(indices=True)
        selected_feature_names = [feature_names[idx] for idx in selected_feature_indices]
        print(selected_feature_names)

    model = model_name
    model.fit(X_train_transformed , y_train)

    train_accuracy = accuracy_score(y_train , model.predict(X_train_transformed))

    y_pred = model.predict(X_test_transformed)
    test_accuracy = accuracy_score(y_test , y_pred)
    precision = precision_score(y_test , y_pred)
    recall = recall_score(y_test , y_pred)
    f1 = f1_score(y_test , y_pred)


    # cross-val-score
    kf = KFold(n_splits=5, shuffle=True)

    # Perform k-fold cross-validation
    scores = cross_val_score(LogisticRegression(max_iter = 500), X_train, y_train, cv=kf)
    cross_val = scores.sum() / 5

    # Create a dictionary with the metrics
    metrics = {
        'Algo Name': [algo_name],
        'train Shape': [X_train_transformed.shape],
        #'features' : [for x in X.columns],
        'Train Size': ['80%'],
        'Train Accuracy': [train_accuracy],
        'Test Accuracy': [test_accuracy],
        'Precision': [precision],
        'Recall': [recall],
        'F1 Score': [f1],
        'cross_val_score' :[cross_val]}
    # Create a DataFrame from the metrics dictionary
    metrics_df = pd.DataFrame(metrics)

    return metrics_df
```

```python
model1 = train_model(LogisticRegression(max_iter =500) ,StandardScaler() ,'LogisticRegression|standarscale' ,X =X_resampled ,y=y_resampled  ,target_col_name='Status'
                     ,feature_selection =None,random_state= 42 )
model1
```

```python
from sklearn.feature_selection import RFE
RFE = train_model(LogisticRegression(max_iter =500) ,StandardScaler() ,'LogisticRegression|standarscale'
                     ,X =X_resampled ,y=y_resampled  ,target_col_name='Status' ,
                     feature_selection =RFE(LogisticRegression(max_iter = 500) , n_features_to_select=20) )
RFE
```

```python
from sklearn.feature_selection import SequentialFeatureSelector
FFS = train_model(LogisticRegression(max_iter =500) ,StandardScaler() ,'LogisticRegression|forward Feature selection' ,X =X_resampled ,y=y_resampled  ,target_col_name='Status'
                     ,feature_selection =SequentialFeatureSelector(LogisticRegression(max_iter =500),direction = 'forward'))
FFS
```

```python
BFS = train_model(LogisticRegression(max_iter =500) ,StandardScaler() ,'LogisticRegression|backward Feature selection' ,X =X_resampled ,y=y_resampled  ,target_col_name='Status'
                     ,feature_selection =SequentialFeatureSelector(LogisticRegression(max_iter =500),direction = 'backward'))
BFS
```

```python
from sklearn.feature_selection import SelectFromModel

embedded_method = train_model(LogisticRegression(max_iter =500) ,StandardScaler() ,'LogisticRegression|Embedded Feature selection' ,X =X_resampled ,y=y_resampled  ,target_col_name='Status'
                     ,feature_selection =SelectFromModel(LogisticRegression(max_iter =500)))
embedded_method
```
