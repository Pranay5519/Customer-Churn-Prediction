## Bivariate

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot 1: Histogram with KDE
sns.histplot(data=df, x='AGE', hue='Status', kde=True, multiple='stack', ax=axes[0])
axes[0].set_title('Distribution of Age by Churn Status')

# Plot 2: Bar Plot of Age Group by Churn Status
age_group = pd.cut(df['AGE'], bins=[0, 18, 30, 40, 50, 60, 100], labels=['0-18', '19-30', '31-40', '41-50', '51-60', '61+'])
sns.countplot(x=age_group, hue='Status', data=df, ax=axes[1])
axes[1].set_title('Bar Plot of Age Group by Churn Status')

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot 1: Stacked Histogram by Churn Status
sns.histplot(data=df, x='CUS_Gender', hue='Status', multiple='stack', ax=axes[0])
axes[0].set_title('Stacked Histogram of CUS_Gender by Churn Status')

# Plot 2: Count Plot of CUS_Gender
sns.countplot(x='CUS_Gender', data=df, ax=axes[1])
axes[1].set_title('Count Plot of CUS_Gender')

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

```python
# check if someone with yearwithus ==15is Churn or not
yearswithus_15 = df[df['YEARS_WITH_US']>=15]
print(yearswithus_15.shape)
sns.countplot(x = yearswithus_15['Status'] , hue = yearswithus_15['CUS_Gender']) # find they are churn...??
```

```python
gender_percentage = df['CUS_Gender'].value_counts(normalize=True) * 100

# Calculate the percentage of churn among males and females
churn_percentage_by_gender = df.groupby('CUS_Gender')['Status'].value_counts(normalize=True).unstack() * 100

# Display the results
print("Percentage of Male and Female Customers:")
print(gender_percentage)

print("\nPercentage of Churn among Males and Females:")
print(churn_percentage_by_gender)
```
